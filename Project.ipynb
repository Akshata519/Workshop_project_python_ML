```python
#rename
data = data.rename(columns = {'Sr No' : 'S_number'})
data.head()
```

```python
#replace
data['startup'] = data['startup'].replace("#Fame" , "Rooman")

```

```python
data['investors'] = data['investors'].replace("Not disclosed" , "TCS")
```

```python
data['investors'] = data['investors'].replace("Undisclosed" , "Tata")
```

```python
data["investors"].value_counts()
```

```python
data["startup"] = data['startup'].str.strip()
```

```python
data["amount"] = data["amount"].astype(int)
```

```python
data['investor'] = data['investors'].fillna('ABC')
```

```python
data["amount"] = data["amount"].fillna(data["amount"].median())
```

```python
data['amount'] = data['amount'].replace({0 : 100})
```

```python
#data visualization using matplotlib library
#how has startup funding trend changed over the year?
import matplotlib.pyplot as plt

demo = data.groupby('year')['amount'].sum()

plt.plot(demo.index, demo.values, marker="o", markerfacecolor="red", color="black", markeredgecolor="green", linestyle="--", linewidth=1.5, label="Total Amount")
plt.title("Startup Funding Trend Over the Years",color='olive',fontsize=16)
plt.xlabel("Year",color='olive',fontsize=12)
plt.ylabel("Total Amount",color='olive',fontsize=12)
plt.grid(linestyle=':',color='blue',linewidth=1.5)
plt.legend()
plt.show()
```

```python
from matplotlib import lines
#which city has highest number of startup
import matplotlib.pyplot as plt


demo = data["city"].value_counts().head(10)
color = ["#FF84BA","#FFDF82","#99C2FF","#FFEFE3","#FF9E20","#BD4444","#73976A","#BBDC12","#FF80C7","#17433F"]

plt.bar(demo.index,demo.values,color = color,edgecolor = 'black')
plt.xticks(rotation=45)
plt.title("Top 10 startup companies in India",color='olive',fontsize=16)
plt.xlabel("City",color='Blue',fontsize=12)
plt.ylabel("Number of Startups",color='Blue',fontsize=12)

plt.show()
```

```python
#which startup sector have the highest number of funded startups
import matplotlib.pyplot as plt

demo = data['vertical'].value_counts().head(10)
color = ["#FF84BA","#FFDF82","#99C2FF","#FFEFE3","#FF9E20","#BD4444","#73976A","#BBDC12","#FF80C7","#17433F"]

plt.figure(figsize=(8,6))
plt.barh(demo.index[::-1],demo.values[::-1],color = color,edgecolor = 'black')

plt.title("Top 10 funded startups",color='olive',fontsize=16)
plt.xlabel("Verticle",color='Blue',fontsize=12)
plt.ylabel("Number of Startups",color='Blue',fontsize=12)

plt.show()
```

```python
#4) what is the distribution of different funding rounds?
import matplotlib.pyplot as plt

demo = data['round'].value_counts().head(5)

plt.pie(demo.values, labels=demo.index, autopct='%1.1f%%')

plt.title("Top 7 different funded rounds",color='olive',fontsize=16)
```

```python
#5).How are startup funding amounts distribution?
import matplotlib.pyplot as plt

demo = data['amount'].value_counts().head()
demo

plt.hist(demo.values, bins=10,edgecolor = "Black")
plt.title("distribution of startup funding")
```

```python
# 6)the realtionship between Funding year and funding Amount?
import matplotlib.pyplot as plt
plt.figure(figsize=(5,6))
plt.scatter(data["year"],data["amount"],color="Red",alpha=0.4,edgecolor="black")
plt.title("Funding amount by year",color='olive',fontsize=16)
plt.xlabel("Year",color='Blue',fontsize=12)
plt.ylabel("Amount",color='Blue',fontsize=12)
plt.grid(linestyle=':',color='Blue',linewidth=1.5)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
sns.countplot(x='round', data=data, order=data['round'].value_counts().head(10).index)

plt.title("Frequency of Different Funding Rounds")
plt.xlabel("Funding Round")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

demo = data["round"].value_counts().head(10).index
sns.countplot(data,y="round",order=demo,palette="viridis")

plt.title("Frequency of Different Funding Rounds")
plt.xlabel("Count")
plt.ylabel("Funding Round")
plt.show()
```
